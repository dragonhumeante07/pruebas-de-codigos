```
import time
import random

# Configuración inicial
n_datos = int(input("Cantidad de datos: "))
lista = [random.randint(0, n_datos) for _ in range(n_datos)]
n = len(lista)

# Inicio de medición de rendimiento
start_time = time.process_time()

# Algoritmo Bubble Sort (Complejidad O(n²))
for i in range(n):
    for j in range(n - i - 1):
        if lista[j] > lista[j+1]:
            # Intercambio de posiciones (Swap)
            lista[j], lista[j+1] = lista[j+1], lista[j]

# Fin de medición y salida de resultados
end_time = time.process_time()

print(f"Lista ordenada: {lista}")
print(f"Duración: {end_time - start_time} segundos")
```

#Burbuja mejorada

Esta versión del algoritmo de burbuja está optimizada. Incluye una bandera (`swapped`) que se establece en `True` si se realiza algún intercambio de elementos durante una pasada. Si una pasada completa finaliza sin intercambios, significa que la lista ya está ordenada y el algoritmo puede terminar antes, mejorando la eficiencia en escenarios de mejor caso o casi ordenados.

```
import time
import random

print("Ingrese la cantidad de datos a ingresar para la burbuja mejorada:")
aleatoreo_mejorada = int(input())

lista_mejorada = []
for _ in range(aleatoreo_mejorada):
    lista_mejorada.append(random.randint(0, aleatoreo_mejorada))

n = len(lista_mejorada)
start_time = time.process_time()
for i in range(0,n):
    swapped = False
    for j in range(0, n - i - 1):
        if lista_mejorada[j] > lista_mejorada[j + 1]:
            lista_mejorada[j], lista_mejorada[j + 1] = lista_mejorada[j + 1], lista_mejorada[j]
            swapped = True


    if not swapped:
        break
end_time = time.process_time()

print(f"Lista ordenada: {lista_mejorada}")
print(f"Tiempo de procesamiento: {end_time - start_time} segundos")
```

```
import time
import random

# Entrada de datos y generación de lista aleatoria
print("Ingrese la cantidad de datos a ingresar para la burbuja mejorada:")
aleatoreo_mejorada = int(input())

lista_mejorada = [random.randint(0, aleatoreo_mejorada) for _ in range(aleatoreo_mejorada)]
n = len(lista_mejorada)

# Inicio del cronómetro de CPU
start_time = time.process_time()

# Algoritmo Burbuja Mejorado
for i in range(n):
    swapped = False  # Bandera para detectar si hubo intercambios en esta pasada

    for j in range(n - i - 1):
        if lista_mejorada[j] > lista_mejorada[j + 1]:
            # Intercambio de valores
            lista_mejorada[j], lista_mejorada[j + 1] = lista_mejorada[j + 1], lista_mejorada[j]
            swapped = True  # Hubo un cambio, la lista aún podría estar desordenada

    # Optimización: Si no hubo cambios, la lista ya está ordenada y salimos del bucle
    if not swapped:
        break

# Fin del cronómetro
end_time = time.process_time()

# Salida de resultados
print(f"Lista ordenada: {lista_mejorada}")
print(f"Tiempo de procesamiento: {end_time - start_time} segundos")
```
