#grafico de la burbuja

```
import time
import random
import matplotlib.pyplot as plt

def bubble_sort(lista):
    n = len(lista)
    for i in range(n):
        for j in range(0, n - i - 1):
            if lista[j] > lista[j+1]:
                lista[j], lista[j+1] = lista[j+1], lista[j]

# Tamaños de prueba
tamaños = [100, 200, 500, 1000, 2000,5000,10000]
tiempos = []

for n in tamaños:
    lista = [random.randint(0, n) for _ in range(n)]
    start = time.process_time()
    bubble_sort(lista)
    end = time.process_time()
    tiempos.append(end - start)

# Graficar
plt.plot(tamaños, tiempos, marker='o')
plt.xlabel("Tamaño de la lista (n)")
plt.ylabel("Tiempo de ejecución (segundos)")
plt.title("Bubble Sort: tiempo vs tamaño de lista")
plt.grid(True)
plt.show()
```

#Burbuja mejorada

Esta versión del algoritmo de burbuja está optimizada. Incluye una bandera (`swapped`) que se establece en `True` si se realiza algún intercambio de elementos durante una pasada. Si una pasada completa finaliza sin intercambios, significa que la lista ya está ordenada y el algoritmo puede terminar antes, mejorando la eficiencia en escenarios de mejor caso o casi ordenados.

```
import time
import random

print("Ingrese la cantidad de datos a ingresar para la burbuja mejorada:")
aleatoreo_mejorada = int(input())

lista_mejorada = []
for _ in range(aleatoreo_mejorada):
    lista_mejorada.append(random.randint(0, aleatoreo_mejorada))

n = len(lista_mejorada)
start_time = time.process_time()
for i in range(0,n):
    swapped = False
    for j in range(0, n - i - 1):
        if lista_mejorada[j] > lista_mejorada[j + 1]:
            lista_mejorada[j], lista_mejorada[j + 1] = lista_mejorada[j + 1], lista_mejorada[j]
            swapped = True


    if not swapped:
        break
end_time = time.process_time()

print(f"Lista ordenada: {lista_mejorada}")
print(f"Tiempo de procesamiento: {end_time - start_time} segundos")
```

#grafico de burbuja mejorada


```
import time
import random
import matplotlib.pyplot as plt

def bubble_sort_mejorada(lista):
    n = len(lista)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if lista[j] > lista[j + 1]:
                lista[j], lista[j + 1] = lista[j + 1], lista[j]
                swapped = True
        if not swapped:
            break

# Lista de tamaños a probar
tamaños = [100, 200, 500, 1000, 2000,5000,10000]
tiempos = []

for n in tamaños:
    lista = [random.randint(0, n) for _ in range(n)]
    start = time.process_time()
    bubble_sort_mejorada(lista)
    end = time.process_time()
    tiempos.append(end - start)

# Graficar
plt.plot(tamaños, tiempos, marker='o')
plt.xlabel("Tamaño de la lista (n)")
plt.ylabel("Tiempo de ejecución (segundos)")
plt.title("Burbuja mejorada: tiempo vs tamaño de lista")
plt.grid(True)
plt.show()
```
