```python
import copy  # Importa la librería para clonar objetos en memoria

class Jugador:
    """Clase que representa los datos del personaje"""
    def __init__(self, posicion=(0, 0), inventario=None, misiones=None):
        self.posicion = posicion  # Guarda la ubicación actual como una Tupla (x, y)
        self.inventario = inventario if inventario is not None else {}  # Crea el inventario como un Diccionario
        self.misiones = misiones if misiones is not None else []  # Crea las misiones como una Lista

    def __str__(self):
        # Devuelve los datos del jugador en un texto legible para la consola
        return f"Posición: {self.posicion} | Inventario: {self.inventario} | Misiones: {self.misiones}"

class Sistema_juego:
    """Clase que controla el motor del juego y su historial"""
    def __init__(self):
        self.estado_actual = Jugador()  # Crea un jugador nuevo en su estado inicial
        self.historial = []  # Pila (Stack) para guardar el pasado (Deshacer)
        self.historial_reacer = []  # Pila (Stack) para guardar lo revertido (Rehacer)
        self.eventos = []  # Cola (Queue) para almacenar eventos en orden cronológico

    def __str__(self):
        return f"Estado Actual: {self.estado_actual}"  # Muestra los datos del jugador actual

    def actualizar_historial(self):
        """Respalda el estado antes de hacer un cambio"""
        copia = copy.deepcopy(self.estado_actual)  # Clona al jugador de forma exacta e independiente
        self.historial.append(copia)  # Guarda la copia en la pila de historial (Push)
        self.historial_reacer.clear()  # Borra el camino de rehacer al iniciar una nueva acción

    def mover_jugador(self, movimiento):
        """Cambia la posición del personaje"""
        self.actualizar_historial()  # Guarda el estado antes de moverlo
        self.estado_actual.posicion = movimiento  # Asigna las nuevas coordenadas
        print(f"[Acción] Jugador se movió a {self.estado_actual.posicion}")

    def avanzar_mision(self, mision):
        """Añade un nuevo objetivo al juego"""
        self.actualizar_historial()  # Guarda el estado antes de modificarlo
        if mision not in self.estado_actual.misiones:  # Filtro: Verifica que la misión no esté repetida
            self.estado_actual.misiones.append(mision)  # Agrega la misión al final de la lista
            print(f"[Acción] Nueva misión registrada: {mision}")

    def completar_mision(self, mision):
        """Busca una misión activa, la marca como completada y la remueve de la lista"""
        if mision not in self.estado_actual.misiones:  # Verifica si la mision no esta activa
            print(f"[Aviso] No se puede completar: '{mision}' no es una misión activa.")
            return

        self.actualizar_historial()  # Guarda el estado en el pasado antes de modificarlo
        self.estado_actual.misiones.remove(mision)  # Elimina la misión de la lista de misiones activas
        print(f"[Acción] ¡Misión completada con éxito!: {mision}")

    def deshacer(self):
        """Vuelve un paso al pasado (Ctrl + Z)"""
        if not self.historial:  # Comprueba si la pila del pasado está vacía
            print("[Historial] No hay acciones para deshacer.")
            return

        copia_actual = copy.deepcopy(self.estado_actual)  # Clona el estado presente
        self.historial_reacer.append(copia_actual)  # Envía el presente a la pila de rehacer

        self.estado_actual = self.historial.pop()  # Extrae el último estado del pasado y lo vuelve el presente (Pop)
        print(f"[Deshacer] Se revirtió la acción")

    def rehacer(self):
        """Vuelve un paso al futuro deshecho (Ctrl + Y)"""
        if not self.historial_reacer:  # Comprueba si hay algo que se pueda restaurar
            print("[Historial] No hay acciones para rehacer.")
            return

        copia_actual = copy.deepcopy(self.estado_actual)  # Clona el estado presente
        self.historial.append(copia_actual)  # Envía el presente de vuelta al historial normal

        self.estado_actual = self.historial_reacer.pop()  # Recupera el estado que se había deshecho (Pop)
        print(f"[Rehacer] Se restableció la acción")

    def modificar_inventario(self, item, cantidad):
        """Añade, suma o quita objetos del inventario"""
        self.actualizar_historial()  # Guarda el estado antes de modificar el inventario
        inventario = self.estado_actual.inventario  # Acceso directo al diccionario de ítems

        if item in inventario:  # Si el ítem ya existe en las llaves del diccionario:
            inventario[item] += cantidad  # Suma o resta la cantidad indicada
            if inventario[item] <= 0:  # Si el objeto se agotó por completo:
                del inventario[item]  # Borra la clave del diccionario para liberar memoria
        elif cantidad > 0:  # Si el objeto es nuevo y la cantidad es positiva:
            inventario[item] = cantidad  # Crea el nuevo ítem con su cantidad inicial

        print(f"[Acción] Inventario actualizado: {item} ({cantidad:+} unidades)")

    def nuevo_evento(self, evento):
        """Añade tareas a la cola de eventos"""
        self.eventos.append(evento)  # Encola el evento al final de la lista (Enqueue)
        print(f"[Cola] Evento recibido: '{evento}'")

    def procesar_siguiente_evento(self):
        """Ejecuta los eventos en estricto orden de llegada (FIFO)"""
        if not self.eventos:  # Comprueba si la cola está vacía
            print("[Cola] No hay eventos pendientes por procesar.")
            return

        evento = self.eventos.pop(0)  # Saca y retorna el primer elemento de la lista (Dequeue)
        print(f"[Procesando] Ejecutando evento: '{evento}'")

        if "pocion" in evento.lower():  # Si el texto contiene la palabra "pocion" (sin importar mayúsculas):
            self.modificar_inventario("Poción", 1)  # Agrega una poción al inventario
        elif "mover" in evento.lower():  # Si el texto contiene la palabra "mover":
            self.mover_jugador((5, 5))  # Mueve automáticamente al jugador al punto (5,5)


game = Sistema_juego ()
print(game)
x = int(input ("ingrese el movimiento en x: "))
y = int(input ("ingrese el movimiento en y: "))
movimiento= (x,y)
game.mover_jugador(movimiento)
game.modificar_inventario("Poción", 3)
game.nuevo_evento("Mover jugador al checkpoint (5,5)")
x = int(input ("ingrese el movimiento en x: "))
y = int(input ("ingrese el movimiento en y: "))
movimiento= (x,y)
game.mover_jugador(movimiento)
game.modificar_inventario("Espada", 1)
game.avanzar_mision("Derrotar al Dragón")
print(game)
game.deshacer()
print(game)
game.rehacer()
print(game)
game.procesar_siguiente_evento()
game.procesar_siguiente_evento()



```
